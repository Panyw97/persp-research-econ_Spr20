```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv('total.csv')
df_2 = pd.read_csv('sentiment.csv')
```

```python
df_1.drop('Unnamed: 0', axis=1, inplace=True)
df_2.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df_3 = df_2[['Issuer', 'wordcount', 'Positive', 'Negative', 'Uncertain', 'Weak_Modal', 'Strong_Modal']]
df_total = df_1.merge(df_3, on='Issuer', how='left')
df_total
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 6))
fig.suptitle('Figure 1. Boxplot of Number of Words & Google Trend', size=15)
ax1 = fig.add_subplot(1, 2, 1)
df_4 = df_2[df_2.wordcount <= 300000]
ax1 = sns.boxplot(x="underpriced", y="wordcount", data=df_4, fliersize=3.5, notch=True)
ax1.set_ylabel('Number of Words')
# ax1.set_xticks([], minor=False)
# plt.legend(loc=3)
ax2 = fig.add_subplot(1, 2, 2)
ax2 = sns.boxplot(x="underpriced", y="avg_trend", data=df_1, fliersize=3.5, notch=True)
ax2.set_ylabel('Google Trend')

fig.savefig('boxplot.jpg')
```

```python
industry_df = df_1[['underpriced', 'Industry']].groupby('Industry').count()
industry_df = industry_df.sort_values(by=['underpriced'], ascending=False)
top_20_industry = industry_df.iloc[:20, 0]
top_ind_df = pd.DataFrame(top_20_industry)
top_ind_df['industry'] = top_ind_df.index
top_ind_df.rename(columns = {'underpriced':'total'}, inplace = True)
```

```python
industry_under = df_1[['underpriced', 'Industry']].groupby('Industry').sum()
top_industry_under = industry_under.sort_values(by=['underpriced'], ascending=False)
top_industry_under['industry'] = top_industry_under.index
top_ind = top_ind_df.merge(top_industry_under, on='industry', how='left')
top_ind
```

```python
plt.figure(figsize=(15,8))
x = np.arange(20)
y1 = top_ind.total
y2 = top_ind.underpriced
bar_width = 0.35
tick_label = top_ind.industry
y3 = y2 /y1
font1 = {'family': 'Times New Roman',
         'weight': 'normal',
         'size': 15,
         }
font2 = {'family': 'Times New Roman',
         'weight': 'normal',
         'size': 12,
         }
plt.bar(x, y1, bar_width, align="center", color="c", label="Total Number", alpha=0.5)
plt.bar(x+bar_width, y2, bar_width, color="b", align="center", label="Underpriced Number", alpha=0.5)
plt.xlabel("Industry", size=12)
plt.ylabel("Number", size=12)
plt.xticks(x-(2 * bar_width), tick_label, rotation=60)
plt.legend(prop=font2)

plt.twinx()

plt.plot(x, y2 / y1, label='Proportion of Underpriced IPO')
plt.xlabel("Industry", size=12)
plt.ylabel("Proportion", size=12)
plt.legend(loc=4, prop=font2)

plt.title('Figure 2. Underpriced IPO Industry', size=20)
plt.savefig('industry.jpg', bbox_inches = 'tight')
```

```python
df2 = df_1[['underpriced', 'Year']].groupby(['Year']).sum()
df3 = df_1[['underpriced', 'Year']].groupby(['Year']).count()
df4 = df2.underpriced / df3.underpriced
df4
```

```python
from matplotlib.pyplot import MultipleLocator
fig = plt.figure(figsize=(10, 5))
fig.suptitle('Figure 3. Number of IPOs and Proportion of Underpriced IPOs in Years', size=15)
ax1 = fig.add_subplot(111)
ax1.plot(df3, color='red', label='Number of IPOs')
ax1.set_ylabel('Number of IPOs')
plt.legend(loc=3)
ax2 = ax1.twinx()  # this is the important function
ax2.plot(df4, color='green', label='Proportion of Underpriced IPOs')
ax2.set_ylabel('Proportion of Underpriced IPOs')

x_major_locator=MultipleLocator(5)
ax1.xaxis.set_major_locator(x_major_locator)

# plt.plot(df3, color='red', label='Number of IPOs')
# plt.plot(df4, color='blue', label='Proportion of Underpriced IPOs')
plt.legend(loc=4)
plt.savefig('IPOs.jpg')
```
